# Final Project: Boston Housing Dataset Analysis

Statistical analysis including descriptive stats, hypothesis testing, correlation, and regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
```

```python
url = 'https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv'
df = pd.read_csv(url)
df.head()
```

## Descriptive Statistics

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure()
sns.histplot(df['medv'], kde=True)
plt.title('Distribution of House Prices')
plt.show()
```

## Correlation Analysis

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.scatterplot(x='rm', y='medv', data=df)
plt.title('Rooms vs Price')
plt.show()
```

## Hypothesis Testing

```python
corr, p_value = stats.pearsonr(df['rm'], df['medv'])
corr, p_value
```

If p-value < 0.05, reject H0.

## Regression Analysis

```python
X = df[['rm']]
y = df['medv']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
sns.regplot(x='rm', y='medv', data=df)
plt.title('Regression Line')
plt.show()
```

## Conclusion
- More rooms → higher price
- Strong correlations exist
- Model confirms relationship
